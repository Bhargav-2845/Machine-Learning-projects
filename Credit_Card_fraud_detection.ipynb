```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
# checking the number of missing values

df.isnull().sum()
```

```python
# distribution of legit and fraudulent transactions

df['Class'].value_counts()
```

# The dataset is highly imbalanced
0 -> normal transaction ,
1 -> fraud transaction

```python
# separating data for analysis

legit =df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data 

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both the transactions

df.groupby('Class').mean()
```

# Under Sampling
Build a sample dataset containing simlar distribution of normal and fraudulent transactions

```python
legit_sample = legit.sample(n = 492)

# conat two dataframes

new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# Spliting the data

```python
X = new_dataset.drop(columns = 'Class',axis = 1)
y = new_dataset['Class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size = 0.2,random_state = 2,stratify =y
)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model Training

Logistic Regression Model

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

# Model Evaluation 

```python
# Accuracy Score for training data

X_train_pred = model.predict(X_train)
train_acc = accuracy_score(X_train_pred,y_train)
print(train_acc)
```

```python
# Accuracy Score for test data

y_pred = model.predict(X_test)
test_acc = accuracy_score(y_pred,y_test)
print(test_acc)
```
